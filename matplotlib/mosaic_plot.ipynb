```python
import numpy as np
import matplotlib.pyplot as plt

# Create 3 datasets from 3 distributions; gaussian, poisson and uniform

# Gaussian distribution
gaussian = np.random.normal(0, 1, 1000)

# Poisson distribution
poisson = np.random.poisson(20, 1000)

# Uniform distribution
uniform = np.random.uniform(-1, 1, 1000)

# Create a figure and axis
fig, ax = plt.subplot_mosaic([['gaussian', 'poisson'], ['uniform', 'uniform']])
ax['gaussian'].hist(gaussian, bins=30, color='b', alpha=0.5)
ax['poisson'].hist(poisson, bins=30, color='r', alpha=0.5)
ax['uniform'].hist(uniform, bins=30, color='g', alpha=0.5)

plt.show()
```

```python
# It's also possible to create a figure which spans an entire column
fig, ax = plt.subplot_mosaic([['gaussian', 'poisson'], ['gaussian', 'uniform']])
ax['gaussian'].hist(gaussian, bins=30, color='b', alpha=0.5)
ax['poisson'].hist(poisson, bins=30, color='r', alpha=0.5)
ax['uniform'].hist(uniform, bins=30, color='g', alpha=0.5)

plt.show()
```

```python
# Example: create a 3x3 grid of subplots
fig, ax = plt.subplot_mosaic([['a', 'b', 'c'], ['d', 'e', 'f'], ['g', 'h', 'i']])
for i in ax:
    if i == 'e':
        continue
    ax[i].hist(np.random.normal(0, 1, 1000), bins=30, color='b', alpha=0.5)
    
plt.tight_layout()
```
